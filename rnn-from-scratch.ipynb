```python
import numpy as np
import random as rand
import math
```

```python
dataset = ["<START> {x} <END>".format(x=x) for x in ["I like football", "I work at Google", "I study in university"]]
```

```python
vocab = [x.split (" ") for x in dataset]
vocab = set (str(x) for x in np.concatenate(vocab))

# vocab size
N = len (vocab)
print (vocab)
```

```python
one_hot = {x: [[0] if i != list(vocab).index (x) else [1] for i in range (len (vocab))] for x in vocab}

for i, j in one_hot.items ():
    print (i, j, sep=(' ' * 20))
```

```python
def mat_add (a, b):
    assert len (a) == len (b) and len (a) and len (a[0]) == len (b[0]), "Invalid matrix sizes {0} and {1}".format (len (a[0]), len (b[0]))
    return [[a[i][j] + b[i][j] for j in range (len(a[0]))] for i in range (len (a))]

def mat_sub (a, b):
    assert len (a) == len (b) and len (a) and len (a[0]) == len (b[0]), "Invalid matrix sizes {0} and {1}".format (len (a[0]), len (b[0]))
    return [[a[i][j] - b[i][j] for j in range (len(a[0]))] for i in range (len (a))]

def mat_mul (a, b):
    assert len (a[0]) == len (b), "Invalid matrix sizes {0} and {1}".format (len (a[0]), len (b))
    
    ar = len (a)
    bc = len (b[0])
    br = len (b)
    
    c = [[0 for _ in range (bc)] for _ in range (ar)]
    
    for i in range (ar):
        for j in range (bc):
            for k in range (br):
                c[i][j] += a[i][k] * b[k][j]
                
    return c

def tanh (a):
    for i in range (len (a)):
        for j in range (len (a[0])):
            a[i][j] = math.tanh (a[i][j])
            
    return a

def softmax (a):
    assert len (a[0]) == 1, "N x 1 matrix required"
    
    b = [[0] for _ in range (len (a))]
    s = 0
    
    for i in range (len (a)):
        s += math.e ** a[i][0]
    
    for i in range (len (a)):
        b[i][0] = (math.e ** a[i][0]) / s
        
    return b

def transpose (a):
    b = [[0 for _ in range (len (a))] for _ in range (len (a[0]))]
    
    for i in range (len (a)):
        for j in range (len (a[0])):
            b[j][i] = a[i][j]
            
    return b
```

```python

# take first sentence
sample = dataset[0].split (" ")

print (sample)

# hidden size
H = 16

# set weights
U = []
V = []
W = []
b = []
c = []
        
for i in range (H):
    U.append ([])
    for j in range (N):
        U[-1].append (rand.random ())
        
for i in range (N):
    V.append ([])
    for j in range (H):
        V[-1].append (rand.random ())
        
for i in range (H):
    W.append ([])
    for j in range (H):
        W[-1].append (rand.random ())
        
for i in range (H):
    b.append ([rand.random ()])
    
for i in range (N):
    c.append ([rand.random ()])
 
print ("Weight U:")
for i in U:
    print (i)

print ("Weight V:")
for i in V:
    print (i)
    
print ("Weight W:")
for i in W:
    print (i)
    
print ("bias b:")
for i in b:
    print (i)
    
print ("bias c:")
for i in c:
    print (i)
```

```python
# one single epoch

h = [[[0] for _ in range (H)]]
x = lambda p: one_hot[sample[p]]

y_t_caps = []

# we start from <START> token
# here we use h(0)
# with h(0) we compute the next h_t and append
# it to h
for t in range (len (sample) - 1):
    # print (x (t))
    
    a_t = mat_add (
        b, mat_add (
            mat_mul (W, h[t]), # h[t] is basically h(t-1)
            mat_mul (U, x(t))
        )
    ) # b + Wh(t - 1) + Ux(t)
    
    h_t = tanh (a_t) # this is the actual h_t
    
    h.append (h_t)
    
    o_t = mat_add (
        c, mat_mul (
            V, h_t
        )
    ) # c + Vh(t)
    
    y_t_cap = softmax (o_t)
    
    # print (y_t_cap)
    y_t_caps.append (y_t_cap)

y_true = [x (i) for i in range (len (sample))]

loss = 0
for i in range (len (y_t_caps)):
    ytr = y_true[i + 1].index ([1]) # we are predicting the next word
    prob = y_t_caps[i][ytr][0]
    
    loss += -1 * math.log (prob + 1e-12)

print (loss)

for i in y_t_caps:
    print (i)
  
print ('=' * 200)  
for i in h:
    print (i)
    
    
    
```
