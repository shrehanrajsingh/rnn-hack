```python
import numpy as np
import random as rand
import math
```

```python
dataset = [
    "<START> I like football <END>",
    "<START> I like cricket <END>",
    "<START> I like basketball <END>",

    "<START> I play football <END>",
    "<START> I play cricket <END>",

    "<START> I work at Google <END>",
    "<START> I work at Microsoft <END>",

    "<START> I study mathematics <END>",
    "<START> I study physics <END>"
]
```

```python
vocab = [x.split (" ") for x in dataset]
vocab = set (str(x) for x in np.concatenate(vocab))

# vocab size
N = len (vocab)
print (vocab)
```

```python
one_hot = {x: [[0] if i != list(vocab).index (x) else [1] for i in range (len (vocab))] for x in vocab}

for i, j in one_hot.items ():
    print (i, j, sep=(' ' * 20))
```

```python
def mat_add (a, b):
    assert len (a) == len (b) and len (a) and len (a[0]) == len (b[0]), "Invalid matrix sizes {0} and {1}".format (len (a[0]), len (b[0]))
    return [[a[i][j] + b[i][j] for j in range (len(a[0]))] for i in range (len (a))]

def mat_sub (a, b):
    assert len (a) == len (b) and len (a) and len (a[0]) == len (b[0]), "Invalid matrix sizes {0} and {1}".format (len (a[0]), len (b[0]))
    return [[a[i][j] - b[i][j] for j in range (len(a[0]))] for i in range (len (a))]

def mat_mul (a, b):
    assert len (a[0]) == len (b), "Invalid matrix sizes {0} and {1}".format (len (a[0]), len (b))
    
    ar = len (a)
    bc = len (b[0])
    br = len (b)
    
    c = [[0 for _ in range (bc)] for _ in range (ar)]
    
    for i in range (ar):
        for j in range (bc):
            for k in range (br):
                c[i][j] += a[i][k] * b[k][j]
                
    return c

def tanh (a):
    for i in range (len (a)):
        for j in range (len (a[0])):
            a[i][j] = math.tanh (a[i][j])
            
    return a

def softmax (a):
    assert len (a[0]) == 1, "N x 1 matrix required"
    
    b = [[0] for _ in range (len (a))]
    s = 0
    
    for i in range (len (a)):
        s += math.e ** a[i][0]
    
    for i in range (len (a)):
        b[i][0] = (math.e ** a[i][0]) / s
        
    return b

def transpose (a):
    b = [[0 for _ in range (len (a))] for _ in range (len (a[0]))]
    
    for i in range (len (a)):
        for j in range (len (a[0])):
            b[j][i] = a[i][j]
            
    return b

def elem_wise_mul (a, b):
    assert len (a) == len (b) and len (a) and len (a[0]) == len (b[0]), "Invalid matrix sizes {0}x{1} and {2}x{3}".format (
        len (a), len (a[0]), 
        len (b), len (b[0])
    )
    
    c = [[0 for _ in range (len (a[0]))] for _ in range (len (a))]
    
    for i in range (len (a)):
        for j in range (len (a[0])):
            c[i][j] = a[i][j] * b[i][j]
    
    return c
```

```python
# take first sentence
sample = dataset[0].split (" ")

print (sample)

# hidden size
H = 16

# set weights
U = []
V = []
W = []
b = []
c = []
        
for i in range (H):
    U.append ([])
    for j in range (N):
        U[-1].append (rand.uniform (-0.1, 0.1))
        
for i in range (N):
    V.append ([])
    for j in range (H):
        V[-1].append (rand.uniform (-0.1, 0.1))
        
for i in range (H):
    W.append ([])
    for j in range (H):
        W[-1].append (rand.uniform (-0.1, 0.1))
        
for i in range (H):
    b.append ([rand.uniform (-0.1, 0.1)])
    
for i in range (N):
    c.append ([rand.uniform (-0.1, 0.1)])
 
print ("Weight U:")
for i in U:
    print (i)

print ("Weight V:")
for i in V:
    print (i)
    
print ("Weight W:")
for i in W:
    print (i)
    
print ("bias b:")
for i in b:
    print (i)
    
print ("bias c:")
for i in c:
    print (i)
```

```python
def one_single_epoch (sample, _print_data = True):
    global U, V, W, b, c
    # one single epoch

    h = [[[0] for _ in range (H)]]
    x = lambda p: one_hot[sample[p]]

    y_t_caps = []

    # we start from <START> token
    # here we use h(0)
    # with h(0) we compute the next h_t and append
    # it to h
    for t in range (len (sample) - 1):
        # print (x (t))
        
        a_t = mat_add (
            b, mat_add (
                mat_mul (W, h[t]), # h[t] is basically h(t-1)
                mat_mul (U, x(t))
            )
        ) # b + Wh(t - 1) + Ux(t)
        
        h_t = tanh (a_t) # this is the actual h_t
        
        h.append (h_t)
        
        o_t = mat_add (
            c, mat_mul (
                V, h_t
            )
        ) # c + Vh(t)
        
        y_t_cap = softmax (o_t)
        
        # print (y_t_cap)
        y_t_caps.append (y_t_cap)

    y_true = [x (i) for i in range (len (sample))]

    loss = 0
    for i in range (len (y_t_caps)):
        ytr = y_true[i + 1].index ([1]) # we are predicting the next word
        prob = y_t_caps[i][ytr][0]
        
        loss += -1 * math.log (prob + 1e-12)

    if _print_data:
        print (loss)

        for i in y_t_caps:
            print (i)
        
        print ('=' * 200)  
        for i in h:
            print (i)
    
    # weight and bias gradient update
    lr = 0.001 # learning rate

    """
        First we will store dL/dh through time
        We know h(t) connects to o(t) and h(t + 1)
        dL/dh = transpose (V)(y_pred - y) + dL/dh(t+1) * dh(t+1)/dh
        we know, dh(t+1)/dh = diag (1 - h(t + 1) ** 2) * W
    """

    dldht = [0 for _ in range (len (y_t_caps))]
    dldht[-1] = mat_mul (
        transpose (V),
        mat_sub (
            y_t_caps[-1],
            y_true[-1]
        )
    )

    for i in range (len (y_t_caps) - 2, -1, -1):
        # ytcaps[i] uses h[i + 1]
        # therefore when we need dldht of i, we need tanh derivative of h[i + 2]
        # which is diag (1 - (h[i + 2])**2)
        hh = [1 - h[i + 2][j][0] ** 2 for j in range (len (h[i + 2]))]
        
        hhh = [[0 for _ in range (H)] for _ in range (H)]
        for j in range (H):
            for k in range (H):
                if j == k:
                    hhh[j][k] = hh[j]
        
        dldht[i] = mat_add (
            mat_mul (
                transpose (V),
                mat_sub (
                    y_t_caps[i],
                    y_true[i + 1]
                )
            ),
            mat_mul (
                mat_mul (
                    transpose (W),
                    hhh
                ),
                dldht[i + 1]
            )
        )
        
    if _print_data:
        for i in dldht:
            print (i)
    
    """
        Update b
        We need to calculate dL/db
        
        dL/db = sum over t (dldht element_wise_mul (1 - (h(t))**2))
        b <- b - lr * dL/db
    """

    dldb = 0

    for t in range (len (y_t_caps)):
        tanh_grad = [[1 - h[t + 1][j][0]**2] for j in range (H)]
        t1 = elem_wise_mul (
            dldht[t],
            tanh_grad
        )
        
        if dldb == 0:
            dldb = t1
        else:
            dldb = mat_add (dldb, t1)

    for i in range (H):
        b[i][0] -= lr * dldb[i][0]

    if _print_data:
        print ("bias b after update:")
        for i in b:
            print (i)
    
    """
        Update c
        We need to calculate dL/dc
        
        dL/dc = sum over t (y_pred (t) - y_true (t))
    """

    dldc = 0

    for i in range (len (y_t_caps)):
        t = mat_sub (y_t_caps[i], y_true[i + 1])
        
        if dldc == 0:
            dldc = t
        else:
            dldc = mat_add (dldc, t)

    for i in range (N):
        c[i][0] -= lr * dldc[i][0]

    if _print_data:
        print ("bias c:")
        for i in c:
            print (i)

    """
        Update U
        We need to calculate dL/dU
        
        dL/dU = sum over t ((dldht element_wise_mul (1 - (h(t))**2))) * transpose (x(t))
    """

    # first we calculate sum over all t (dldht element_wise_mul (1 - (h(t) ** 2)))

    dldht_tanhd = []

    for t in range (len (y_t_caps)):
        tanh_grad = [[1 - h[t + 1][j][0]**2] for j in range (H)]
        
        t1 = elem_wise_mul (
            dldht[t],
            tanh_grad    
        )
        
        dldht_tanhd.append (t1)

    dldu = 0

    for t in range (len (y_t_caps)):
        t1 = mat_mul (
            dldht_tanhd[t],
            transpose (x (t))
        )
        
        if dldu == 0:
            dldu = t1
        else:
            dldu = mat_add (dldu, t1)
            
    for i in range (H):
        for j in range (N):
            U[i][j] -= lr * dldu[i][j]

    if _print_data:
        print ("weight U:")
        for i in U:
            print (i)

    """
        Update V
        We need to calculate dL/dV
        
        dL/dV = sum over t (y_pred (t) - y_true (t)) * h(t)
    """

    dldv = 0

    for t in range (len (y_t_caps)):
        t1 = mat_mul (
            mat_sub (
                y_t_caps[t],
                y_true[t + 1]
            ),
            transpose (h[t + 1])
        )
        
        dldv = t1 if dldv == 0 else mat_add (dldv, t1)

    # update V
    for i in range (N):
        for j in range (H):
            V[i][j] -= lr * dldv[i][j]

    if _print_data:
        print ("weight V:")
        for i in V:
            print (i)

    """
        Update W
        We need to calculate dL/dW
        
        dL/dW = sum over t (dL/dh (t) elem_wise_mul (1 - (h(t))**2)) * transpose (h(t - 1))
    """

    dldw = 0

    for t in range (len (y_t_caps)):
        t1 = mat_mul (
            dldht_tanhd[t], # dL/dh(t + 1)
            transpose (h[t]) # h(t)
        )
        
        dldw = t1 if dldw == 0 else mat_add (dldw, t1)

    # update W

    for i in range (H):
        for j in range (H):
            W[i][j] -= lr * dldw[i][j]
        
    if _print_data:
        print ("weight W:")
        for i in W:
            print (i)
```

```python
# Train
NUM_EPOCHS = 10

for i in range (1, NUM_EPOCHS + 1):
    print (f'epoch {i}')
    
    rand.shuffle (dataset)
    dataset *= 2 # increase dataset size each epoch for better learning
    
    for sentence in dataset:
        sample = sentence.split(" ")
        one_single_epoch(sample, _print_data=False)
```

```python
def predict (context: str):
    spl_c = ['<START>', *context.split (' ')]
    
    h_t = [[0] for _ in range (H)]
    
    for tok in spl_c:
        a_t = mat_add (
            b, mat_add (
                mat_mul (
                    W, h_t
                ),
                mat_mul (
                    U, one_hot[tok]
                )
            )
        )
        
        h_t = tanh (a_t)
    
    o_l = mat_add (
        c, mat_mul (V, h_t)
    )
    
    y_pred = softmax (o_l)
    m = 0
    mm = None
    
    for i, iv in enumerate (y_pred):
        if iv[0] > m:
            m = iv[0]
            mm = i
        
        # print (iv)
    
    assert mm != None
    d = [[0] if j != mm else [1] for j in range (N)]
    
    # print ("Probabilities:")
    # for i, iv in enumerate (vocab):
    #     print (iv, y_pred[i])
    
    v = list (vocab)
    return sorted ([(v[i], y_pred[i]) for i in range (len (v))], key=lambda p: p[-1], reverse=True)
    
    
```

```python
# test prediction
test = ["I like", "I work", "I work at"]

for i in test:
    print (f'{i} {predict (i)[:3]}')

print (predict ("I")[:3])
```
